```python
import pandas as pd
import numpy as np
import pickle as pkl

from webStash import WebStash
from bs4 import BeautifulSoup
```

```python
# https://www.findmypast.co.uk/search/results?_page=0&datasetname=1911%20census%20for%20england%20%26%20wales&firstname=alex&firstname_variants=true&lastname=sieusahai
# Is what I'm going to use

def gen_link(first_name, last_name):
    return 'https://www.findmypast.co.uk/search/results?_page=0&datasetname=1911%20census%20for%20england%20%26%20wales&firstname='+first_name+'&firstname_variants=true&lastname='+last_name
```

```python
df = pd.read_csv('name_information.csv')
df.head()
```

```python
# lets figure out if these people were British or not

stash = WebStash(getterType='chromedriver', waitTimeBeforeScraping=5)

amap = {}

for [_, data] in df.iterrows():
    first_name = data['CommonName'].split()[0]
    last_name = data['LastName']
    link = gen_link(first_name, last_name)
    print(link)
    html = stash.get_web_data(link).html
    soup = BeautifulSoup(html, 'lxml')
    while True:
        try:
            result_elem = soup.select('div#result')[0]
            break
        except IndexError:
            print('failed; trying again...')
            del stash.cacher[link]
            html = stash.get_web_data(link).html
            soup = BeautifulSoup(html, 'lxml')
    if len(result_elem.select('h2')):
        amap[data['Name']] = 0
    else:
        amap[data['Name']] = 1

print(amap)
```

```python
pkl.dump(amap, open('name_isBritish_map.pkl', 'wb'))
```
